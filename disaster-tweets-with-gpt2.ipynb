# 3. Process data

Export data to separate variable.

```python
text_cleaned_list = train.text_cleaned.values
text_cleaned_list_test = test.text_cleaned.values
```

We calculate maximum length which is to be used as a fixed input size.

```python
MAX_LENGTH = np.max([len(str(x).split()) for x in text_cleaned_list]) + 5
MAX_LENGTH
```

Define special tokens as Beginning of a String, End of a String, Pad and Unknown token. We use native GPT2 tokenizer to process text.

```python
BOS_TOKEN = "<s>"
PAD_TOKEN = "<pad>"
EOS_TOKEN = "</s>"

tokenizer = GPT2TokenizerFast.from_pretrained("gpt2", max_length=MAX_LENGTH)
tokenizer.add_special_tokens({
    "eos_token": EOS_TOKEN,
    "bos_token": BOS_TOKEN,
    "unk_token": "<unk>",
    "pad_token": PAD_TOKEN,
#     "mask_token": "<mask>"
})
```

Add BOS and EOS token to the beginning and the end of each tweet.

```python
train_data = [BOS_TOKEN + " " + str(ex) + " " + EOS_TOKEN for ex in text_cleaned_list]
test_data = [BOS_TOKEN + " " + str(ex) + " " + EOS_TOKEN for ex in text_cleaned_list_test]
```

```python
train_data[:10]
```

Here we convert tokens to token_id so that each sequence has a fixed length (MAX_LENGTH), which is reached with 'pad_to_max_length=True' parameter.

```python
train_data_ = [tokenizer(str(x), return_tensors='tf', max_length=MAX_LENGTH, truncation=True, pad_to_max_length=True, add_special_tokens=True)['input_ids'] for x in train_data]
train_data_in = tf.squeeze(tf.convert_to_tensor(train_data_), axis=1)

test_data_ = [tokenizer(str(x), return_tensors='tf', max_length=MAX_LENGTH, truncation=True, pad_to_max_length=True, add_special_tokens=True)['input_ids'] for x in test_data]
test_data_in = tf.squeeze(tf.convert_to_tensor(test_data_), axis=1)
```

```python
train_data_in.shape
```

```python
labels = list(train.target.values)
y_train_in = tf.constant(labels, dtype=tf.int32)
```

We must also create a mask for each sequence. Mask specifies the which positions in the sequence is occupied with tokens (1) and which with pad tokens (0).

```python
train_data_mask_ = [tokenizer(str(x), return_tensors='tf', max_length=MAX_LENGTH, truncation=True, pad_to_max_length=True, add_special_tokens=True)["attention_mask"] for x in train_data]
train_data_mask = tf.squeeze(tf.convert_to_tensor(train_data_mask_), axis=1)

test_data_mask_ = [tokenizer(str(x), return_tensors='tf', max_length=MAX_LENGTH, truncation=True, pad_to_max_length=True, add_special_tokens=True)["attention_mask"] for x in test_data]
test_data_mask = tf.squeeze(tf.convert_to_tensor(test_data_mask_), axis=1)
```

# 4. GPT2 model

```python
model = TFGPT2Model.from_pretrained("gpt2", use_cache=False,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id)
model.training = True
model.resize_token_embeddings(len(tokenizer))
```

```python
for layer in model.layers:
    layer.trainable = False
     

model.summary()
```

```python
input = tf.keras.layers.Input(shape=(None,), dtype='int32')
mask = tf.keras.layers.Input(shape=(None,), dtype='int32')
x = model(input, attention_mask=mask)
x = tf.reduce_mean(x.last_hidden_state, axis=1)
x = tf.keras.layers.Dense(16, activation='relu')(x)
x = tf.keras.layers.Dropout(0.3)(x)
output = tf.keras.layers.Dense(1, activation='sigmoid')(x)
     

clf = tf.keras.Model([input, mask], output)
     

clf.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=3e-5,epsilon=1e-08, clipnorm=1.0)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
base_learning_rate = 0.0005
optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate)
#loss=tf.keras.losses.BinaryCrossentropy()
loss=tf.keras.losses.BinaryCrossentropy()

clf.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
     

callbacks = tf.keras.callbacks.EarlyStopping(
        monitor="accuracy", verbose=1, patience=3, restore_best_weights=True)
```

```python
tf.config.experimental_run_functions_eagerly(True)
```

# 5. Train and Predict

```python
history = clf.fit([train_data_in, train_data_mask], y_train_in, epochs=30, batch_size=32, validation_split=0.2, callbacks=callbacks)
```
